```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
sc = spark.sparkContext
```

#partie1


```python
rdd1 = sc.textFile('/content/Shakespeare.txt')
```

```python
rdd1.collect()
```

```python
rdd1.count()
```

```python
words=rdd1.flatMap(lambda x : x.split())
```

```python
words.collect()
```

```python
word_map = words.map(lambda word : (word ,1))
```

```python
word_map.collect()
```

```python
s=word_map.reduceByKey(lambda a ,b : (a+b)) 
```

```python
s.collect()
```

```python
s.coalesce(1).saveAsTextFile("content/")
```
